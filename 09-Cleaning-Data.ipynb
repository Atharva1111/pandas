```python
import pandas as pd
import numpy as np
```

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
df = pd.DataFrame(people)
df
```

```python
# Deletes all the rows with missing value

df.dropna()
```

```python
# The axis can be index or columns that should be deleted if there is a missing value
# The how parameter can be all, any, etc. 

df.dropna(axis='index', how='any')
```

```python
df.dropna(axis='index', how='all')
```

```python
# Delete row if email is missing

df.dropna(axis='index', how='all', subset=['email'])
```

```python
# Deletes row if the last or email is missing because how is set to any

df.dropna(axis='index', how='any', subset=['last', 'email'])
```

```python
# Deletes row if the last and  email is missing because how is set to all

df.dropna(axis='index', how='all', subset=['last', 'email'])
```

```python
# Replace custom missing values

df.replace('NA', np.nan, inplace=True)
df.replace('Missing', np.nan, inplace=True)
df
```

```python
# Check if the value is missig

df.isna()
```

```python
# Fill mising values

df.fillna('MISSING')
```

```python
# Get datatype of columns
# Object is string

df.dtypes
```

```python
# Type of np.nan is float

type(np.nan)
```

```python
# Casting age to float
# Cannot convert to int beacuse np.nan is float

df['age'] = df['age'].astype('float')
df
```

```python
df.dtypes
```

```python
# Average of age

df['age'].mean()
```
