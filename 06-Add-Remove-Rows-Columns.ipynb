```python
import pandas as pd
```

```python
people = {
    "first": ["Corey", 'Jane', 'John'], 
    "last": ["Schafer", 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
df = pd.DataFrame(people)
df
```

```python
# Adding a column

df["full_name"] = df["first"] + " " + df["last"]
df
```

```python
# Remove a column

df.drop(columns=['full_name'], inplace=True)
df
```

```python
# Add a single row

df = pd.concat([df, pd.DataFrame({"first": ["Tony"]})], ignore_index=True)
df
```

```python
avengers = {
    "first": ["Tony", "Steve"],
    "last": ["Stark", "Rogers"],
    "email": ["IronMan@avenge.com", "Cap@avenge.com"]
}
avengers_df = pd.DataFrame(avengers)
avengers_df
```

```python
# Append another DataFrame

df = pd.concat([df, avengers_df], ignore_index=True)
df
```

```python
# Drop a row

df.drop(index=5)
df
```

```python
# Drop rows with a condition

filt = df["last"] == "Doe"
df.drop(index=df[filt].index)
```
