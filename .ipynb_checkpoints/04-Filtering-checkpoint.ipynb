```python
import pandas as pd
```

```python
people = {
    "first": ["Corey", 'Jane', 'John'], 
    "last": ["Schafer", 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
df = pd.DataFrame(people)
```

```python
filt = df['last'] == 'Doe'
filt
```

```python
df[filt]
```

```python
df.loc[filt]
```

```python
df.loc[filt, 'email']
```

```python
filt1 = (df['first'] == 'John') & (df['last'] == 'Doe')
df.loc[filt1, 'email']
```

```python
df.loc[~filt1, 'email']
```
