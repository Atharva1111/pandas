```python
country_grp = df.groupby(['Country'])
country_grp.get_group('India')
```

```python
country_grp['Salary'].median().loc['India']
```

```python
country_grp['Salary'].agg(['median', 'mean']).loc['India']
```

```python
# Cannot use str method because it is SeriesGroupBy object and not Series object
# Instead use apply method

country_uses_python = country_grp['HaveWorkedLanguage'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
country_respondance = df['Country'].value_counts()
country_respondance
```

```python
python_df = pd.concat([country_respondance, country_uses_python], axis='columns')
python_df
```

```python
# Renaming the column

python_df.rename(columns={'Country': 'NumRespondents', 'HaveWorkedLanguage': 'NumKnowsPython'}, inplace=True)
python_df
```

```python
# Get Percentage

python_df['PctKnowsPython'] = (python_df['NumKnowsPython']/python_df['NumRespondents'])*100
python_df
```

```python
# Sort by percentage

python_df.sort_values(by='PctKnowsPython', ascending=False)
```
